```python
#!pip install qiskit
# !pip install pylatexenc
from qiskit import QuantumCircuit
from qiskit.circuit.library import ZZFeatureMap, PauliFeatureMap, TwoLocal
import matplotlib.pyplot as plt
import pylatexenc
from qiskit.visualization import circuit_drawer
```

```python
# ZZFeatureMap of 2 qubits, repeated twice
circuit = ZZFeatureMap(2, reps=2, insert_barriers=True)
decomposed_circuit = circuit.decompose()

image_file = "ZZ_2q_2rep_image.png"
fig = circuit_drawer(decomposed_circuit, scale=0.7, output="mpl")
fig.savefig(image_file, dpi=500)
plt.close(fig)
```

```python
# ZZFeatureMap of 3 qubits, repeated twice
circuit = ZZFeatureMap(2, reps=1, insert_barriers=True)
decomposed_circuit = circuit.decompose()

image_file = "ZZ_2q_1rep_image.png"
fig = circuit_drawer(decomposed_circuit, scale=0.7, output="mpl")
fig.savefig(image_file, dpi=500)
plt.close(fig)
circuit.decompose().draw("mpl")
```

```python
prep = PauliFeatureMap(2, reps=1, paulis=['ZZ'], insert_barriers=True)
prep.decompose().draw("mpl")
```

```python
circuit = ZZFeatureMap(2, reps=2, insert_barriers=True)
circuit.decompose().draw("mpl")
```

```python
twolocal = TwoLocal(num_qubits=3, reps=1, rotation_blocks=['ry','rz'], 
               entanglement_blocks='cx', entanglement='linear', insert_barriers=True)

decomposed_circuit = twolocal.decompose()
image_file = "2_local_3q_1rep_image.png"
fig = circuit_drawer(decomposed_circuit, scale=0.7, output="mpl")
fig.savefig(image_file, dpi=500)
plt.close(fig)
twolocal.decompose().draw('mpl')
```
